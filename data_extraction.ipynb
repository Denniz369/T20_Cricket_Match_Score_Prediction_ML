```python
import pandas as pd
import numpy as np
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir('t20s'):                  #creates a list of files in t20s
    filenames.append(os.path.join('t20s', file)) # adds t20s with the file name to create a filepath
```

```python
filenames[0:5]
```

```python
import pandas as pd
from tqdm import tqdm
from yaml import safe_load  

final_df = pd.DataFrame()    #creating a empty dataframe
counter = 1                  # initiating counter

for file in tqdm(filenames):
    try:
        with open(file, 'r') as f:
            df = pd.json_normalize(safe_load(f))                     #loading each yaml file in tabular format
            df['match_id'] = counter                                 # Adds new columns match_id with the current value
            final_df = pd.concat([final_df, df], ignore_index=True)  # Accumulating all files as a single file
            counter += 1                                             # For every loop new match_id is created
    except:
        print(f"[error INFO] - problem in reading the file: {file}")
        continue

final_df
```

## Now we are creating a data frame in a suitable format(ball by ball) for building a ML model

```python
count = 1
delivery_df = pd.DataFrame()

# Iterate over each row in final_df
for index, row in tqdm(final_df.iterrows(),total = final_df.shape[0],desc = 'Processing rows'):  #iterates through each row of final_df
    count += 1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    
    # Iterate over deliveries in the 1st innings
    for ball in row ['innings'][0]['1st innings']['deliveries']: #extracting ball by ball data
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')

    # Create a DataFrame for the current match's deliveries
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)  # Concatenate the current match's data with the main DataFrame

            
```

```python
delivery_df
```

```python
# This function will extract the team which is not batting ie. "bowling_team"

def bow(row):
    for team in row['teams']:
            if team != row['batting_team']:
                return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bow, axis = 1)  # using bow function we are ceating a bowling team column
```

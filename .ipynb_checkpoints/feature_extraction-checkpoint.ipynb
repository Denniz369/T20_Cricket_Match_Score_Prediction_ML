```python
df = pickle.load(open('df_20_08.pkl', 'rb'))
```

```python
df
```

```python
#There are all city names so venue is not needed

df.drop(['venue'],axis = 1,inplace=True)
df.head()
```

```python
# we are extracting cities where more than 600 matches happened

eligible_cities=df['city'].value_counts()[df['city'].value_counts()>600].index.tolist()
```

```python
#Excluding cities where less than 600 matches happened

df=df[df['city'].isin(eligible_cities)]
```

```python
df.head(20)
```

```python
# Creating a column to show the current score at that ball

df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df.head(50)
```

```python

#Creating a column to show the number of ball on that over

df['ball_no']=df['ball'].apply(lambda x: str(x).split('.')[1])
```

```python
df
```

```python
# Creating column showing number of current over

df['overs']=df['ball'].apply(lambda x: str(x).split('.')[0])
df.head()
```

```python
# Column to show the number of balls bowled

df['balls_bowled'] = df['overs'].astype('int')*6 + df['ball_no'].astype('int')
df.head(20)
```

```python
df[df['match_id']==2]
```

```python
# balls left
df['balls_left'] = 120 - df['balls_bowled']
df.tail(50)
```

```python
# We are encoding the player dismissed column  to transform into a cumulative columns 

df['player_dismissed'] = df['player_dismissed'].apply(lambda x :0 if x=='0' else 1)
df
```

```python
# Transforming the player dismissed column as a column which shows the total number of players dismissed at that ball

df['player_dismissed']=df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df.tail(50)
```

```python
df.tail(10)
```

```python
# balls left
df['wicketss_left'] = 10 - df['player_dismissed']
df
```

```python
# creating current run rate column

df['crr'] = df['current_score']*6 / df['balls_bowled']
```

```python
df
```

```python
# last five over

groups = df.groupby('match_id')

match_ids = df['match_id'].unique()

last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
last_five
```

```python
df['last_five'] = last_five
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on='match_id')
```

```python
final_df
```

```python
# selecting important colums
final_df.drop(columns=['match_id', 'runs_y', 'player_dismissed', 'overs', 'ball_no', 'balls_bowled'],
             inplace = True,axis=1)
```

```python
final_df
```

```python
df.isnull().sum()
```

```python
# we are dropping all the rows with null values and the null values will be from the first five overs of the "last_five" column
# Eventhough other column values are deleted along with them, it does not affect the algorithm because it still shows the performance of the teams clearly

final_df.dropna(inplace = True)
```

```python
final_df
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.compose import ColumnTransformer # allows to apply different preprocessing steps to different columns of a dataset
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False, drop='first'), ['batting_team', 'city','bowling_team'])
], remainder='passthrough')
```

```python
pip install xgboost
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
pipe = Pipeline(steps = [
    ('step_1', trf),
    ('step_2', StandardScaler()),
    ('step_3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_Dept=12, random_state=101))
])
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
r2=r2_score(y_test, y_pred)
mae=(mean_absolute_error(y_test, y_pred))
```

```python
n = len(y_test)
p = X_test.shape[1]
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
```
